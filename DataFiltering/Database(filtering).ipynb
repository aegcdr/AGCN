```python
import pandas as pd
```

```python
mut=pd.read_csv('/content/drive/MyDrive/Datasets/mutation_filtered.csv')
mut.head()
```

```python
mut.rename( columns={'Unnamed: 0':'cell_name'}, inplace=True )
mut.head()
```

```python
meth=pd.read_csv('/content/drive/MyDrive/Datasets/meth_filtered.csv')
meth.head()
```

```python
cnv=pd.read_csv('/content/drive/MyDrive/Datasets/cnv_filtered.csv')
cnv.head()
```

```python
cnv.rename( columns={'Unnamed: 0':'cell_name'}, inplace=True )
cnv.head()
```

```python
exp=pd.read_csv('/content/drive/MyDrive/Datasets/expression_filtered_v2.csv')
exp.head()
```

```python
exp.rename( columns={'Unnamed: 0':'cell_id'}, inplace=True )
exp.head()
```

```python
info=pd.read_csv('/content/drive/MyDrive/Datasets/mutation.csv')
info.head()
```

```python
info=info.sort_values('cell_line_name')
info.head()
```

```python
cell_dict = dict(zip(info.cosmic_sample_id.astype(str), info.cell_line_name))
cell_dict
```

```python

for index, r in exp.iterrows():
    r1=r['cell_id']
    temp=cell_dict[r1.split('.')[1]]
    exp.loc[index, 'cell_id'] = temp
exp.head()
```

```python
exp=exp.sort_values('cell_id')
exp.head()
```

```python
drug=pd.read_csv('/content/drive/MyDrive/Datasets/drug_response_filtered.csv')
drug.head()
```

```python
drug.rename( columns={'Unnamed: 0':'cell_name'}, inplace=True )
drug.head()
```

```python
want=[]
want.append(mut['cell_name'].unique().tolist())
want
```

```python
want.append(meth['cell_name'].unique().tolist())
want.append(cnv['cell_name'].unique().tolist())
want.append(exp['cell_id'].unique().tolist())
want.append(drug['cell_name'].unique().tolist())
want
```

```python
filt=set(want[0]).intersection(*want)
filt
```

```python
len(filt)
```

```python
df_mut=mut.loc[mut['cell_name'].isin(filt)]
df_mut
```

```python
df_meth=meth.loc[meth['cell_name'].isin(filt)]
df_meth
```

```python
df_cnv=cnv.loc[cnv['cell_name'].isin(filt)]
df_cnv
```

```python
df_exp=exp.loc[exp['cell_id'].isin(filt)]
df_exp
```

```python
df_drug=drug.loc[drug['cell_name'].isin(filt)]
df_drug
```

```python
df_mut.to_csv('/content/drive/MyDrive/Datasets/final_mutation.csv')
```

```python
df_meth.to_csv('/content/drive/MyDrive/Datasets/final_meth.csv')
```

```python
df_cnv.to_csv('/content/drive/MyDrive/Datasets/final_cnv.csv')
```

```python
df_exp.to_csv('/content/drive/MyDrive/Datasets/final_expression.csv')
```

```python
df_drug.to_excel('/content/drive/MyDrive/Datasets/final_drug.xlsx')
```
