```python
mut=pd.read_csv("/content/drive/MyDrive/Datasets/mutation.csv")
mut.head()
```

```python
temp=mut[['cell_line_name','genetic_feature','is_mutated']]
temp
```

```python
row=temp['cell_line_name'].unique()
col=temp['genetic_feature'].unique()
col
```

```python
df=pd.DataFrame(index=row)
df[col]=0
```

```python
for index, r in temp.iterrows():
    r1=r['cell_line_name']
    r2=r['genetic_feature']
    r3=r['is_mutated']
    df.at[r1, r2]=r3

df.head()
```

```python
df2=df.sort_index()
df2
```

```python
df2.to_csv('/content/drive/MyDrive/Datasets/mutation_filtered.csv')
```
