```python
r=temp['CELL_LINE_NAME'].unique()
c=temp['DRUG_NAME'].unique()
len(c)
```

```python
df=pd.DataFrame(index=r,columns=c)
df.head()
```

```python
df=df.sort_index()
df.head()
```

```python
for index, r in temp.iterrows():
    r1=r['CELL_LINE_NAME']
    r2=r['DRUG_NAME']
    r3=r['LN_IC50']
    df.at[r1, r2]=r3

df
```

```python
df = df.fillna(0)
df
```

```python
zeroValues = df.eq(0).mean()
print (zeroValues)
```

```python
df = df.loc[:, zeroValues <= 0.1]
df
```

```python
df.all()
```

```python
for col in df.columns :
    #print(col)
  mean_value=df[col].mean()
  df[col] = df[col].replace(0,mean_value)

#df
```

```python
df.all()
```

```python
df.to_excel('/content/drive/MyDrive/Datasets/103SelectedDrugs.xlsx')
```
