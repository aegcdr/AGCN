```python
import os
import deepchem as dc
from rdkit import Chem
import numpy as np
import hickle as hkl
import pandas as pd


drug_smiles_file='/content/drive/MyDrive/Datasets/103_wanted_smiles.xlsx'
save_dir='/content/drive/MyDrive/Datasets/103_smiles'
df = pd.read_excel('/content/drive/MyDrive/Datasets/103_wanted_smiles.xlsx')
pubchemid2smile = {}

for index, x in df.iterrows():
	#print(x[1])
	pubchemid2smile[x["Name"]] = x["Smiles"]
	#cid = {item.split('\t')[0]}
print(pubchemid2smile)
#pubchemid2smile = {item.split('\t')[0]:item.split('\t')[1].strip() for item in open(drug_smiles_file).readlines()}
#pubchemid2smile = {'Cisplatin': "N.N.Cl[Pt+2]Cl"}

if not os.path.exists(save_dir):
    os.makedirs(save_dir)
molecules = []
print(pubchemid2smile)
# print(cid)
ind=0
keys = list(pubchemid2smile.keys())
for each in keys :
	print(f'{keys.index(each)} : {each}')
	molecules=[]
	molecule = Chem.MolFromSmiles(pubchemid2smile[each])
	molecules.append(molecule)
	featurizer = dc.feat.graph_features.ConvMolFeaturizer()
	print(f'Molecule : {molecules}')
	print(f'Featurizer : {featurizer}')
	try :
		mol_object = featurizer.featurize(molecules)
	except :
		print(f'Failed for : {molecule}')
	if not mol_object[0] :
		break
	print(f'MOL Object : {mol_object[0].atom_features}')
	features = mol_object[0].atom_features
	degree_list = mol_object[0].deg_list
	adj_list = mol_object[0].canon_adj_list
	#hkl.dump([features,adj_list,degree_list],'%s/%s.hkl'%(save_dir,each))
print(adj_list)
```
