```python
df = df[['bcr_patient_barcode','pharmaceutical_therapy_drug_name','treatment_best_response']]
```

```python
df = df[1:]
```

```python
df['bcr_patient_barcode'] = df['bcr_patient_barcode'].astype(str)+'-01'
```

```python
for i in df.index:
    if df.loc[i]['treatment_best_response'] in ['Stable Disease', 'Clinical Progressive Disease']:
        df.loc[i]['treatment_best_response']=1
    elif df.loc[i]['treatment_best_response'] in ['Complete Response', 'Partial Response']:
        df.loc[i]['treatment_best_response']=0
```
